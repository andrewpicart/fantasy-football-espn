```python
# create df for league
league_2016 = pd.DataFrame()
for d in range(len(league.teams)):
    temp_df = pd.DataFrame.from_dict(league.teams[d].__dict__)
    league_2016 = league_2016.append(temp_df)
```

```python
league_2016
```

```python
# add year (season) to the league
league_2016['year'] = 2016
```

```python
# change the index to 'week'
league_2016.index.names = ['week']
```

```python
# map according to draft order
# no easy way to do this but manually
league_2016['draft_order'] = league_2016['team_id'].map({5:1, 10:2, 9:3, 2:4, 12:5, 3:6, 6:7, 7:8, 4:9, 11:10, 8:11, 1:12})
```

```python
# use this help with draft order
# ordered in list of team_id
league_2016['owner'].unique()
```

```python
# drop unecessary columns
league_2016 = league_2016.drop(['division_id', 'team_id', 'team_abbrev', 'team_name', 'streak_length', 'streak_type', 'logo_url', 'standing', 'roster', 'outcomes', 'schedule'], axis=1)
```

```python
# wildcard winner in playoffs
# must do manually if the 6th place is not automatically the wild card winner
league_2016['playoffs'] = np.where(league_2016['final_standing'] == 7, 1, league_2016['final_standing'])
```

```python
# playoffs category for all other teams
league_2016['playoffs'] = np.where(league_2016['final_standing'] <= 5, 1, 0)
```
