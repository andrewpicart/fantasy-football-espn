## Modeling

I think in order for modeling to work, I need to make a df from the current week of past seasons. For example, if it's week 5 in 2019, I need to make a df of everything before week 6 from past_seasons. It would be very easy to tell what the final placements are from the end of the season.

```python
from sklearn import linear_model
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LassoCV, RidgeCV
from sklearn.model_selection import train_test_split
```

```python
X = past_regular_seasons.drop(['final_standing'], axis=1)
y = past_regular_seasons['final_standing']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
lr = linear_model.LinearRegression()
```

```python
sc = StandardScaler()
```

```python
X_train = sc.fit_transform(X_train)
```

```python
lasso = LassoCV()
ridge = RidgeCV()
```

```python
lr.fit(X_train, y_train)
```

```python
lasso.fit(X_train, y_train)
ridge.fit(X_train, y_train)
```

```python
lr.score(X_train, y_train)
```

```python
lr.score(X_test, y_test)
```

```python
lasso.score(X_train, y_train)
```

```python
lasso.score(X_test, y_test)
```

```python
ridge.score(X_train, y_train)
```

```python
ridge.score(X_test, y_test)
```
